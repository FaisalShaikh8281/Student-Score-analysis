# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x='Gender')
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gp,annot=True)
plt.title('Releation between parents education and student score')
plt.show()
```

# From the above chart is concluded that the education of the parents have the good impact on students scores

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gp1, annot= True)
plt.title("Releation between parents MAritial status and student score")
plt.show()
```

# From the above chart we have concluded there is no impact on the students scores due to the parents martial status

# Distribution of Ethnic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "Group B", "group C", "group D","group E"]
mlist = [groupA["EthnicGroup"].sum(), groupB["EthnicGroup"].sum(), groupC["EthnicGroup"].sum(), groupD["EthnicGroup"].sum(), groupE["EthnicGroup"].sum()]
print(mlist)
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethenic groups")
plt.show()
```
